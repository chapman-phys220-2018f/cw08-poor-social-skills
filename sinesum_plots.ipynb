## Approximations of a Sine Sum Function

In the following notebook, the implementation and results of a given sine summation are demonstrated.  The formula for the summation is given by
$$S_n(t) = \frac{4}{\pi}\sum_{k=1}^n \frac{1}{2k-1} \sin\left(\frac{2(2k-1)\pi t}{T}\right)$$
where t and T are user-defined variables.  The summation is known to converge toward a value given by a secondary function
$$f(t) = \begin{cases} 1, & 0 < t < T/2 \\ 0, & t = 0 \\ -1, & -T/2 < t < 0 \end{cases}$$
where t and T are the same variables used in $S_n(t)$

The implementation used to compute $S_n(t)$ contains three variables: $T$, $t$, and $n$. The variable $n$ is used as the upper bound of the summation, and tells the program how many times to run before returning the sum.  In this module, the variable $T$ is held constant as $T = \frac{2}{\pi}$.  The inputs for $t$ are then defined as percentages of this number; specifically, $t = \alpha T$ where $\alpha = 0.01, 0.25, 0.49$.  The values of $n$ are then varied, to demonstrate convergence as $n \xrightarrow{} \infty$.

```python
# Defining a few constants and helpre functions. The xvals function will save some space as the n value changes, and plotVals makes a reusable graph function.
T = 2 / np.pi
def xvals(n): return np.arange(1, n+1)
def plotVals(x, y, l):
    f = plt.figure(figsize = (12,8))
    a = plt.axes()
    if (len(x) > 1):
        a.plot(x, y, color="red", label=l)
    else:
        a.scatter(x, y, color="red", label=l)
    a.set(xlabel="n", ylabel="Sum", ylim=(0,2), title="Graph of {}".format(l))
    plt.show()
```

```python
alpha = 0.01
n = 1
S1 = ss.s(T, alpha*T, n)
x = xvals(n)
plotVals(x, S1, "Alpha = {}, n = {}".format(alpha, n))
```

```python
alpha = 0.01
n = 3
x = xvals(n)
S1 = np.array([ss.s(T, alpha*T, p) for p in x])
plotVals(x, S1, "Alpha = {}, n = {}".format(alpha, n))
```

#### Graph Set 1

In the initial graph, the value for alpha was set to be $0.01$, and subsequently $n$ was allowed to vary over a range from 1 to 100.  For $n=1$ and $n=3$, the approximation of the function is quite poor, with $n=1$ returning only a single point and $n=3$ resulting in a line.  At $n=30$, a slight curve is seen and the end of the graph begins to converge toward the correct value.  With $n=100$, the graph can be clearly seen to be converging toward 1.  Obviously, as the value of $n$ increases, the sum computed becomes more and more accurate.  It is interesting that the function converges as a continuous curve instead of a sawtooth-looking function as is usually seen.

#### Graph Set 3

In an interesting turn of events, the final set of graphs are nearly identical to the first set.  This implies that placing the $t$ value directly at $\frac{T}{4}$ has the greatest degree of effect on the function, while values closer to the boundaries effect the graph less.  The "effective" alpha value could possibly be formulated as 
$$\alpha_{real} = \frac{T}{4} - \Big(\Big|\frac{T}{4} - \alpha\Big|\Big)$$
which would account for $0.01$ and $0.49$ displaying the same behavior.  However, further testing is needed to determine if this relationship holds for all values of alpha.  Additionally, the current set of graphs only tests values using positive $t$.  Perhaps negative values for $t$ would display a different interaction.
